```python
# now we will see the samples whose kraken information is not available

kraken_missing = Df[Df["Kraken_total"].isna() | (Df["Kraken_total"] == 0)]
print("\n All samples missing Kraken info (NaN or 0):")
kraken_missing
```

```python
# Now we will keep only those samples in which greater than or equal to 80 percent of the reads are classified
# This will also remove the above 3  sample with NaN


# Step 1: Identify rows below 80%
below_80 = Df[Df["Classified_%"] < 80]

# Step 2: Identify rows >= 80%
above_equal_80 = Df[Df["Classified_%"] >= 80]


print(" Samples with <80% classified:")
print(below_80)

# Step 3: Keep only >=80% rows
Df_filtered = above_equal_80.copy()

print(" Filtered dataframe shape:", Df_filtered.shape)
Df_filtered.shape
```

```python
Df_filtered
```

```python
# now we will check the overall statistics of this table

Df_filtered.info()
Df_filtered.describe()
```

```python
# Main Filtration
# --> Each sample should contain atleast 1 million reads for the effective classification by kraken

import pandas as pd


# Filter: keep only samples with Kraken_total >= 1,000,000
Df_final = Df_filtered[Df_filtered["Kraken_total"] >= 1_000_000].copy()

print("Number of samples with >=1M reads in Kraken_total:")
Df_final
```

```python
discarded = Df_filtered[Df_filtered["Kraken_total"] < 1_000_000]
print(" Samples with <1M reads in Kraken_total:")
print(discarded)
```

```python
# Since all the samples have greater than 1 million reads non of them were discarded

# --> Classified reads Per Bioproject = average reads in millions


# Group by BioProject and calculate mean of Classified_reads
avg_classified_per_project = (
    Df_final.groupby("BioProject")["Classified_reads"]
    .mean()                  # average reads
    / 1_000_000  )            # convert to millions


print("Average Classified Reads per BioProject (in millions):")
print(avg_classified_per_project)
```

Check no. of samples remained, Average no. of classified and unclassifeied reads for each bioprojects.

```python
# After filtration
Samples_per_Project = Df_final.groupby("BioProject")["Kraken_total"].count()
print(Samples_per_Project)
```

```python
import pandas as pd


# First, create Total_reads column (Classified + Unclassified)
Df_final["Total_reads"] = Df_final["Classified_reads"] + Df_final["Unclassified_reads"]

# Group by BioProject and calculate required stats
project_stats = Df_final.groupby("BioProject").agg(
    Total_Reads_Million=("Total_reads", lambda x: round(x.mean()/1_000_000, 2)),
    Avg_Classified_Reads_Million=("Classified_reads", lambda x: round(x.mean()/1_000_000, 2)),
    Avg_Unclassified_Reads_Million=("Unclassified_reads", lambda x: round(x.mean()/1_000_000, 2))
).reset_index()

print(" Reads per BioProject (in millions):")
project_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Dataset 1: Reads per BioProject
data_reads = project_stats

# Dataset 2: Number of samples per BioProject
data_samples = Samples_per_Project

# Merge the two datasets on the 'BioProject' column
df_merged = pd.merge(data_reads, data_samples, on='BioProject')

# Sort the merged DataFrame by Total_Reads_Million in descending order
df_merged = df_merged.sort_values(by='Total_Reads_Million', ascending=True)

# Create the labels for the y-axis
df_merged['Label'] = df_merged['BioProject'] + ' (' + df_merged['Kraken_total'].astype(str) + ')'

# Generate the horizontal stacked bar chart
fig, ax = plt.subplots(figsize=(12,6))

# Plot the classified reads
p1 = ax.barh(df_merged['Label'], df_merged['Avg_Classified_Reads_Million'], color='green', label='Classified')

# Plot the unclassified reads, stacked on top of the classified reads
p2 = ax.barh(df_merged['Label'], df_merged['Avg_Unclassified_Reads_Million'], left=df_merged['Avg_Classified_Reads_Million'], color='red', label='Unclassified')

# Add percentage labels to the bars
for index, row in df_merged.iterrows():
    total = row['Total_Reads_Million']
    classified_pct = (row['Avg_Classified_Reads_Million'] / total) * 100
    unclassified_pct = (row['Avg_Unclassified_Reads_Million'] / total) * 100

    # Position the percentage label for classified reads (green bar)
    ax.text(row['Avg_Classified_Reads_Million'] * 0.5, row['Label'], f'{classified_pct:.0f}%', ha='center', va='center', color='white', fontsize=8)

    # Position the percentage label for unclassified reads (red bar)
    ax.text(row['Avg_Classified_Reads_Million'] + row['Avg_Unclassified_Reads_Million'] * 0.5, row['Label'], f'{unclassified_pct:.0f}%', ha='center', va='center', color='white', fontsize=8)

# Set chart title and labels
ax.set_title('Average proportion of classified vs. unclassified reads per BioProject ', pad=20)
ax.set_xlabel('Average Reads (Millions)')
ax.set_ylabel('BioProject (Number of Samples)')
ax.legend(loc='upper right', bbox_to_anchor=(0.95, 1.05))
plt.tight_layout()

# Save the plot
plt.savefig('bio_project_stacked_bar.png')
```

```python
import pandas as pd
import numpy as np
from io import StringIO


df = Df_final.copy()

# Define a function to extract the sample name from the 'File' column
def extract_sample_name(file_string):
    if isinstance(file_string, str):
        parts = file_string.split('_')
        if len(parts) > 3:
            sample_name = parts[3].split('.')[0]
            if sample_name.startswith('SRR') or sample_name.startswith('ERR'):
                return sample_name
    return np.nan

# Add the 'common' column to the DataFrame
df['common'] = df['File'].apply(extract_sample_name)

# Display the resulting DataFrame
print("Resulting DataFrame with the new 'common' column:")
df
```
